```python
from dbrepo.api.dto import (
    QueryDefinition,
    JoinDefinition,
    ConditionalDefinition,
    JoinType
)

DATABASE_ID = "a6229b2e-5a3d-4a46-93aa-e9474f492ec1"

query = QueryDefinition(
    datasources=[
        "coffee_consumption",
        "sleep_metrics"
    ],
    columns=[
        "coffee_consumption.participant_id",
        "coffee_consumption.coffee_intake_cups_per_day",
        "coffee_consumption.caffeine_mg_per_day",
        "sleep_metrics.sleep_hours",
        "sleep_metrics.sleep_quality"
    ],
    joins=[
        JoinDefinition(
            type=JoinType.INNER,
            datasource="sleep_metrics",
            conditionals=[
                ConditionalDefinition(
                    column="coffee_consumption.participant_id",
                    foreign_column="sleep_metrics.participant_id"
                )
            ]
        )
    ]
)

view = client.create_view(
    database_id=DATABASE_ID,
    name="coffee_sleep_view",
    query=query,
    is_public=True,
    is_schema_public=True
)

print(view)
```

```python
from dbrepo.api.dto import QueryDefinition, FilterDefinition, FilterType

query = QueryDefinition(
    datasources=["coffee_consumption"],
    columns=[
        "coffee_consumption.participant_id",
        "coffee_consumption.coffee_intake_cups_per_day",
        "coffee_consumption.caffeine_mg_per_day"
    ],
    filters=[
        FilterDefinition(
            type=FilterType.WHERE,
            column="coffee_consumption.coffee_intake_cups_per_day",
            operator=">=",
            value="4"
        )
    ]
)

subset = client.create_subset(
    database_id=DATABASE_ID,
    query=query
)

subset.head()
```
